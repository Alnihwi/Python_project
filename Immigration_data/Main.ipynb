```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('D:\\Immigration_data\Canada.xlsx')
df.head()
```

```python
df.columns
```

```python
df2 = df.loc[: ,["AreaName"] + list(df.columns[9:])]
df2.columns
```

```python
df2.set_index("AreaName", inplace = True)
df2
```

```python
df2["Total"] = df2.sum(axis = 1)
df2
```

```python
df2 = df2.groupby("AreaName").sum()
df2
```

```python
df2["Total"].plot(
    kind="pie",
    figsize=(15, 6),
    autopct="%1.1f%%",
    labels = None,
    shadow=True
)


plt.title("Immigration to Canada from 1980 - 2013")
plt.axis("equal")
plt.legend(labels = df2.index, loc="upper left")
plt.savefig("Pie_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df_years = df2.drop(columns=["Total"]).T


plt.figure(figsize=(12,6))

plt.plot(df_years.index, df_years.sum(axis=1), marker="o")

plt.title("Immigration to Canada (1980 - 2013)")
plt.xlabel("Year")
plt.ylabel("Total")

plt.grid(True)
plt.show()


years = df2.drop(columns=["Total"]).columns
values = df2.drop(columns=["Total"]).sum()

plt.plot(years, values)

plt.savefig("Linegraph.png", dpi=300, bbox_inches="tight")
plt.show()
```
